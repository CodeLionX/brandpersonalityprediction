## Setup plotly

You need the following codeblock in each notebook to get plotly running offline and smothly:

```python
# import plotly components for offline display of diagrams
from IPython.display import display, HTML
import plotly.offline as offline
from plotly.graph_objs import *


# enable latex-mathsymbols in plotly
offline.init_notebook_mode(connected=True)

# The polling here is to ensure that plotly.js has already been loaded before
# setting display alignment in order to avoid a race condition.
display(HTML(
    '<script>'
        'var waitForPlotly = setInterval( function() {'
            'if( typeof(window.Plotly) !== "undefined" ){'
                'MathJax.Hub.Config({ SVG: { font: "STIX-Web" }, displayAlign: "center" });'
                'MathJax.Hub.Queue(["setRenderer", MathJax.Hub, "SVG"]);'
                'clearInterval(waitForPlotly);'
            '}}, 250 );'
    '</script>'
))

# general imports
import pandas as pd
import numpy as np
import datetime
```

## Read and prepare data

```python
df = pd.read_csv('mypersonality_final.csv', encoding = "ISO-8859-1")
print(df)
```

## Plot 3D Scatter

```python
trace = Scatter3d(
    x = df['sEXT'],
    y = df['sNEU'],
    z = df['sAGR'],
    name = '',
    mode='markers',
    marker = dict(opacity=0.2, size=8)
)
layout = Layout(
    title = 'First three traits',
    scene = dict(
        xaxis = dict(type='lin', title='Extraversion'),
        yaxis = dict(type='lin', title='Neuroticism'),
        zaxis = dict(type='lin', title='Agreeableness')
    )
)
fig = dict(data = [trace], layout=layout)
offline.iplot(fig)
```

# Demo for a radar chart

```python
trace = Scatter(
    t = ['EXT', 'NEU', 'AGR', 'CON', 'OPN'],
    r = [df['sEXT'][0], df['sNEU'][0], df['sAGR'][0], df['sCON'][0], df['sOPN'][0]],
    mode='lines+markers',
    name='test',
    marker=dict(
        color='blue',
        size=100,
        line=dict(
            color='blue',
            width=2
        ),
        opacity=0.7
    )
)
layout = Layout(
    title='Big Five Traits',
    font=dict(
        size=15
    )
)
fig = dict(data = [trace], layout=layout)
offline.iplot(fig)
```
